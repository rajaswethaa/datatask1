```python
import pandas as pd
import numpy as np

# STEP 1 : LOAD DATASET

df = pd.read_csv("/content/drive/MyDrive/ultimate_student_productivity_dataset_5000.csv")

print("First 10 rows:")
print(df.head(10))

print("\nDataset Shape:")
print(df.shape)

print("\nColumn Names:")
print(df.columns)
```

```python

# STEP 2 : DATASET INFO

print("\nDataset Info:")
print(df.info())

print("\nData Types:")
print(df.dtypes)
```

```python

# STEP 3 : SUMMARY STATS

print("\nSummary Statistics:")
print(df.describe())
```

```python

# STEP 4 : CHECK MISSING VALUES

print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# -------------------------
# STEP 6 : REMOVE DUPLICATES
# -------------------------
df = df.drop_duplicates()

print("\nShape After Removing Duplicates:")
print(df.shape)
```

```python
# -------------------------
# STEP 7 : UNIQUE VALUES
# -------------------------
print("\nGender Values:")
print(df["gender"].unique())

print("\nAcademic Levels:")
print(df["academic_level"].unique())

print("\nInternet Quality:")
print(df["internet_quality"].unique())
```

```python

# STEP 8 : FEATURE ENGINEERING


# Age group
df["age_group"] = pd.cut(
    df["age"],
    bins=[15,18,21,25,30],
    labels=["Teen","Young","Adult","Senior"]
)

# Total study time
df["total_study_time"] = df["study_hours"] + df["self_study_hours"]
```

```python

# STEP 9 : BASIC ANALYSIS


print("\nAverage Productivity by Gender:")
print(df.groupby("gender")["productivity_score"].mean())

print("\nAverage Exam Score by Academic Level:")
print(df.groupby("academic_level")["exam_score"].mean())

print("\nAverage Sleep Hours:")
print(df["sleep_hours"].mean())
```

```python

# STEP 10 : OUTLIER CHECK


print("\nStudents with very low sleep (<4 hrs):")
print(df[df["sleep_hours"] < 4].head())

print("\nStudents with high social media (>6 hrs):")
print(df[df["social_media_hours"] > 6].head())
```

```python

# STEP 11 : SORT DATA


top_students = df.sort_values(by="exam_score", ascending=False)

print("\nTop 10 Students by Exam Score:")
print(top_students.head(10))
```

```python

# STEP 12 : SAVE CLEAN DATASET


df.to_csv("student_productivity_cleaned.csv", index=False)

print("\nCleaned Dataset Saved Successfully")
```

```python

# STEP 13 : EXPORT SUMMARY


summary = df.describe()

summary.to_csv("dataset_summary.csv")

print("\nSummary File Saved")
```
